```python
from ultralytics import YOLO

# Load a model
model = YOLO('yolov8s.pt')  # load a pretrained model (recommended for training)

# Train the model
results = model.train(data='cfg/dataset/coco.yaml', epochs=20, imgsz=640,verbose=True, exist_ok=True)
```
